Fourier Similarity Theorem:

Given that $$f(x) = FT(g(t))$$

then
$$f(ax) = \frac{1}{|a|} g\left(\frac{t}{a}\right)$$

$$f(ax) = \frac{1}{|a|} \left[ \Sigma_{t=0}^n g(t/a) exp(-2\pi i x t / a)  \right]$$

```python
# sample data
x = arange(10)
g = exp(-((x-5)**2))
plot(x,g)
```

```python
# frequencies (with an extra axis for later)
freq = fftfreq(g.size)[:,newaxis]
```

```python
# output indices - selected to interpolate between some indices we have
# the upsampling factor in this case is 1, because the spacing is the same
outinds = np.array([3.5,4.5,5.5,6.5])[newaxis,:]
```

```python
# the kernel is e^(-2pi i) * our new indices
kern=np.exp((-1j*2*pi)*freq*outinds);
```

```python
# freq has the same size as the input, so we're dotting along that axis
# but we operate on the fft(g), which should have the appropriate freqs since we used fftfreq
# we use the ifft because apparently that's what our kernel definition implies
result = np.dot(ifftn(g),kern)
```

```python
# what do we get?
print result
plot(real(result))
```

```python
# overplot that on the original
plot(g)
plot(outinds.squeeze(), real(result))
```

```python
# what if we want to upsample rather than just shift?
# frequencies don't change
freq = fftfreq(g.size)[:,newaxis]
# output indices do - change our step size to 0.5
outinds = np.arange(0,9,0.5)[newaxis,:]
# same kernel expression
kern=np.exp((-1j*2*pi)*freq*outinds)
# apparently, using the ifft, we never need to divide by a, which doesn't seem right...
result = np.dot(ifftn(g),kern)
```

```python
plot(g)
plot(outinds.squeeze(),result.real)
```

```python
# in principle, we should now be able to zoom a lot... and not even using regular samples
outinds = np.array([2,3,3.25,3.5,4.,4.25,4.5,4.75,5,5.25])[newaxis,:]
kern=np.exp((-1j*2*pi)*freq*outinds)
result = np.dot(ifftn(g),kern)
```

```python
plot(g)
plot(outinds.squeeze(),result.real)
```

```python
# that was pretty good, let's try something more sensible
outinds = np.linspace(2,8,30)[newaxis,:]
kern=np.exp((-1j*2*pi)*freq*outinds)
result = np.dot(ifftn(g),kern)
plot(g)
plot(outinds.squeeze(),result.real)
plot(outinds.squeeze(),exp(-(outinds.squeeze()-5)**2))
```

```python
# not bad!  How far off are they?
plot(outinds.squeeze(),result.real-exp(-(outinds.squeeze()-5)**2))
```

```python
# what if we had picked an asymmetric input?
def experiment(f,x,outinds):
    g = f(x)
    freq = fftfreq(g.size)[:,newaxis]
    if len(outinds.shape) == 1:
        outinds = outinds[newaxis,:]
    kern=np.exp((-1j*2*pi)*freq*outinds)
    result = np.dot(ifftn(g),kern)
    plot(g)
    plot(outinds.squeeze(),result.real)
    plot(outinds.squeeze(),f(outinds.squeeze()))
    figure()
    plot(outinds.squeeze(),result.real-f(outinds.squeeze()))
```

```python
experiment(lambda x: exp(-(x-2)**2), arange(10), linspace(0,10,20))
```

```python
experiment(lambda x: exp(-(x-2)**2)+exp(-(x-8)**2), arange(10), linspace(0,10,150))
```

```python
# this one doesn't work
experiment(lambda x: exp(-(x-2)**2)+exp(-(x-8)**2), linspace(0,10,20), linspace(0,10,150))
```

```python
# this one does
experiment(lambda x: exp(-(x/2.-2)**2)+exp(-(x/2.-8)**2), arange(20), linspace(0,20,150))
```

```python
# How do these compare to interpolation?
import scipy.interpolate
def interpcompare(f,x,outinds):
    g = f(x)
    freq = fftfreq(g.size)[:,newaxis]
    if len(outinds.shape) == 1:
        outinds = outinds[newaxis,:]
    kern=np.exp((-1j*2*pi)*freq*outinds)
    result = np.dot(ifftn(g),kern)
    figure(figsize=[16,10])

    plot(outinds.squeeze(),result.real,label='fourier', marker='o')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='linear',bounds_error=False)(outinds.squeeze()),label='linear')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='slinear',bounds_error=False)(outinds.squeeze()),label='slinear')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='quadratic',bounds_error=False)(outinds.squeeze()),label='quadratic')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='cubic',bounds_error=False)(outinds.squeeze()),label='cubic')
    plot(g,label='input',linestyle="--",marker='s')
    plot(outinds.squeeze(),f(outinds.squeeze()),label='correct',linestyle=":")

    legend(loc='best')
    figure(figsize=[16,10])

    plot(outinds.squeeze(),result.real-f(outinds.squeeze()),label='fourier',marker='o')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='linear',bounds_error=False)(outinds.squeeze())-f(outinds.squeeze()),label='linear',marker='^')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='slinear',bounds_error=False)(outinds.squeeze())-f(outinds.squeeze()),label='slinear')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='quadratic',bounds_error=False)(outinds.squeeze())-f(outinds.squeeze()),label='quadratic')
    plot(outinds.squeeze(),scipy.interpolate.interp1d(x,f(x),kind='cubic',bounds_error=False)(outinds.squeeze())-f(outinds.squeeze()),label='cubic')
    legend(loc='best')
```

```python
# this is more interesting, though: in the presence of noise, how does it do?
def f(x):
    np.random.seed(42)
    return exp(-(x/4.-2.)**2/(2.*(1/3.)**2))+exp(-(x/4.-8.)**2/(2.*(1/5.)**2))+1+np.random.randn(x.size)/4
interpcompare(f, arange(0,40), arange(0,40,0.1))
# this is a failed experiment: the 'correct' has different noise than the 'input' because of shape offsets (despite a consistent seed)n
```

Some accuracy tests:

```python
def gaussian(x):
    return np.exp(-x**2/2.)

def accuracy_test(imsize,upsample_factor):
    from image_registration.fft_tools import upsample
    x = np.arange(imsize)
    g = gaussian(x - (imsize-1)/2.)
    xz,z = upsample.zoom1d(g,upsample_factor,outsize=g.size*upsample_factor,return_xouts=True)

    return  ((gaussian(xz - (imsize-1)/2.)-z)**2).sum()
```

```python
accuracy_of_ups10 = [accuracy_test(10,us) for us in xrange(1,50)]
accuracy_of_ups20 = [accuracy_test(20,us) for us in xrange(1,50)]
accuracy_of_ups30 = [accuracy_test(30,us) for us in xrange(1,50)]
clf()
plot(arange(1,50),accuracy_of_ups10)
plot(arange(1,50),accuracy_of_ups20)
plot(arange(1,50),accuracy_of_ups30)
xlabel('upsample factor')
```

```python
accuracy_of_size = [accuracy_test(s,5) for s in xrange(5,50)]
clf(); 
plot(arange(5,50),accuracy_of_size)
plot(arange(5,50),0.005/arange(5,50)+0.001/5)
```

```python
accuracy_of_size = [accuracy_test(s,10.) for s in xrange(5,50)]
clf(); 
plot(arange(5,50),accuracy_of_size)
plot(arange(5,50),0.02/arange(5,50)+0.001/10)
```

```python
actest = array([[accuracy_test(s,u) for s in xrange(5,105,5)] for u in xrange(1,50,3)])
```

```python
theoretical = array([[4e-5*upsample_factor+0.12*imsize**-1.83 for imsize in xrange(5,105,5)] for upsample_factor in xrange(1,50,3)])
```

```python
def f(x,a,b):
    return a*x**b
from scipy.optimize import curve_fit
curve_fit(f,arange(1,50,3), actest[:,-1])
```
